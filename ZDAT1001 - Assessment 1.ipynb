Coding Task and Question: By looking back at the tutorial on data cleaning, choose a method for cleaning these dirty data values and implement it (you should use a different approach for each). In the next three blocks:
- include your code for this (about 10 lines)
- explain your approaches
- re-plot the chart from earlier to check that your approach has been effective

```python
import numpy as np
sales['SKU14256'].replace(0, np.nan, inplace=True)
sales['SKU14256'].interpolate(method='linear', inplace=True)

max_value = sales['SKU13425'].max()
sales['SKU13425'].replace(max_value, np.nan, inplace=True)
sales.SKU13425.fillna(sales['SKU13425'].median(), inplace=True)
```

In the case of SKU14256, the dirty data was a zero - the only one in the column. The approach was to convert that into a np.nan and then follow with an interpolation for the NaN. A linear interpolation was chosen because in this case it provided a value equal to the midpoint of the two datapoints before and after the NaN. This fit well with the look of the line chart and avoided use of the mean, which would have been skewed by the extreme peak and trough around periods 20-25. An ffil, bfill or median could also have been used here.

A median value was used to replace the spike in SKU13425. This was chosen because, as mentioned previously, the mean is skewed by the outlier. The back fill option would not work as well here because the previous value was a zero. A linear interpolation and forward fill would also work here.
